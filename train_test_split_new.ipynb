```python
import pandas as pd 
from helper_functions import helpers
from helper_functions import clustering_helpers

import warnings
warnings.filterwarnings("ignore")
RSEED = 2

```

```python
demographics = pd.read_hdf('C:/Users/Maria/Desktop/Work/Projects/MIMIC/Representation-Clustering/patient_clustering_EHR/data/data_icu.h5', key='demographics') 
#C:/Users/Maria/Desktop/Work/Projects/MIMIC/Representation-Clustering/patient_clustering_EHR/patient_clustering_EHR/
labs_vitals = pd.read_hdf('C:/Users/Maria/Desktop/Work/Projects/MIMIC/Representation-Clustering/patient_clustering_EHR/labs_vitals.h5', key = 'preprocessed') 

interventions = pd.read_hdf('C:/Users/Maria/Desktop/Work/Projects/MIMIC/Representation-Clustering/patient_clustering_EHR/interventions.h5', key='preprocessed') 
```

```python
labs_vitals.drop(['hours_in'],axis=1, inplace=True)
labs_vitals.rename(columns={'hours_in_':'hours_in'}, inplace=True)
labs_vitals.reset_index(inplace=True)
labs_vitals.set_index(['icustay_id', 'subject_id', 'hadm_id', 'hours_in'], inplace=True)
labs_vitals.head()
```

```python
interventions.drop(['hours_in'],axis=1, inplace=True)
interventions.rename(columns={'hours_in_':'hours_in'}, inplace=True)
interventions.reset_index(inplace=True)
interventions.set_index(['icustay_id', 'subject_id', 'hadm_id', 'hours_in'], inplace=True)
interventions.head()
```

```python
icu_series = labs_vitals.join(interventions, how='outer')
icu_series.head()
```

```python
del interventions
del labs_vitals
```

```python

# reading demographics and storing patients sets
icu = icu_series.reset_index()
#storing the true labels
true_labels = demographics[demographics.subject_id.isin(icu.subject_id)][['subject_id', 'hadm_id', 'icustay_id', 'mort_icu']]
del icu 


list_of_features = list(icu_series.columns)

positive = (demographics.mort_icu.value_counts().values[0])
random_sample_positive = int(positive/5)
random_sample_positive_subj_id = demographics[demographics.mort_icu == 0].sample(n=random_sample_positive, random_state=RSEED).subject_id
negative_subj_id = demographics[demographics.mort_icu == 1].subject_id
icu_series.reset_index(inplace=True)
icu_series = icu_series[(icu_series.subject_id.isin(negative_subj_id.values)) | (icu_series.subject_id.isin(random_sample_positive_subj_id.values))]
icu_series.set_index(['subject_id', 'hadm_id', 'icustay_id'], inplace=True)
icu_series.head()
demographics.set_index("subject_id", inplace=True)
```

```python
icu_series.head()
```

```python
from sklearn.model_selection import train_test_split
demo_train, demo_test, y_train, y_test = train_test_split(demographics, demographics.mort_icu, test_size=0.9, random_state=42)
icu_series = icu_series.reset_index()
demo_train = demo_train.reset_index()
demo_test = demo_test.reset_index()

icu_series_train = icu_series[icu_series.subject_id.isin(demo_train.subject_id)]
y_train = y_train[y_train.index.isin(icu_series_train.subject_id)]
icu_series_test = icu_series[icu_series.subject_id.isin(demo_test.subject_id)]
y_test = y_test[y_test.index.isin(icu_series_test.subject_id)]


icu_series_train.to_hdf('train_test_set.h5', key='Xtrain') 
y_train.to_hdf('train_test_set.h5', key='ytrain') 
icu_series_test.to_hdf('train_test_set.h5', key='Xtest') 
y_test.to_hdf('train_test_set.h5', key='ytest') 
```
